```python
# Run this cell without changes
df = pd.read_csv('../pandas-202/diamonds.csv')
df.head()
```

**1) What are the mean carat and the median carat for the each group of color in this dataset?** 

```python
# Code here to find the mean carat and median carat
df2 = df[['carat','color']]
mean_carat = df2.groupby(['color']).mean()

#df.groupby(['Animal']).mean()

mean_carat
```

```python
df2.groupby(['color']).median()
```

In your own words, how are the mean and median related to each other and what do these values tell us about the distribution of the variable 'Age'

# Your written answer here
The higher carat diamonds have similar mean and median values which suggests a tighter distribution for these particular colors. 

The lower carat diamonds have more variance between mean and median, specifically, the median values are notably lower than the mean values. This suggests a greater distribution and outliers with high carat values for these color categories.
